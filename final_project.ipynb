### The COVID-19 virus emerged in Wuhan, China in November of 2019. It was declared a pandemic in March of 2020.
```python
df = pandas.read_excel('s3://gersony-ia241/covid_data.xls')
```

## **Question 1: Which Countries Suffered the Highest Death Rates (As of Dec 14, 2020)** 
### Per Capita COVID-19 Deaths -- Top 25 Countries

```python
df['per_capita_deaths']=df['deaths']/df['popData2019']
sum_deaths_per_capita_by_country = df.groupby('countriesAndTerritories').sum()['per_capita_deaths']
sum_deaths_per_capita_by_country.nlargest(25).plot.bar()
```

### The bar chart displays the 25 countries with the highest per capita COVID death rates at the time the data was published. Latin American and European countries are represented prominently in this list. Belgium and San Marino reported per capita COVID deaths that outpaced other listed countries by a significant margin. The United States is the most populous country on the list. 


## **Question 2: Within the Country with the Highest Per Capita COVID-19 Death Rate, How Were the Deaths Distributed Over Time?**
### Isolating Belgium Data

```python
belgium_data = df.loc[df['countriesAndTerritories'] == 'Belgium']
```

### Mapping Deaths onto Timeline

```python
belgium_data.plot.line(x='dateRep',y='deaths')
```

### This line graph depicts Belgian COVID deaths by date within the scope of the data. While COVID deaths persist throughout the relevant period, significant surges are noted in April and November. The distribution of the deaths is not uniform, but rather dominated by periods of elevated mortal risk followed by more calm periods. 


## **Question 3: How Does Population Size Correlate Nationally with Per Capita COVID-19 Deaths?** 
### Are (Demographically) Smaller or Larger Countries Most Affected by COVID-19?

```python
df.plot.scatter(x='popData2019', y='per_capita_deaths')
```

### Population size is negatively correlated with per capita COVID deaths in this period. Most of the countries with the highest per capita COVID death tolls are well under 50 million in population, and several microstates are featured prominently as states with high population-adjusted death tolls from COVID-19. Across most of the distribution, however, outliers exist, as some countries with lower populations experienced lower death rates. Several larger countries, such as the United States, Italy, France, and Spain, experienced high death rates. Some countries appear to display negative per capita death rates, indicating limited data entry mistakes.
